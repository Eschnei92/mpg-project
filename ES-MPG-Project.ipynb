```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


mpgal = sns.load_dataset("mpg")
sns.set(style="darkgrid")
```

```python
#Create a visual showing the relationship between horsepower and fuel economy.

sns.lineplot(
    data=mpgal, 
  x="horsepower", y="mpg",
    color='green',
    linewidth=2
    )
plt.title("MPG change relative to horsepower")
```

```python
#Create a table summarizing fuel economy by country of origin

mpg_data = sns.load_dataset('mpg')

summ_table = mpg_data.groupby('origin')['mpg'].agg(['mean', 'median']) 

summ_table['mean'] = summ_table['mean'].round(1)  

print("MPG by Country of Origin")  
print(summ_table)
```

```python
#Create a visual showing how fuel economy has changed over time
sns.lineplot(
    data=mpgal, 
  x="model_year", y="mpg",
    color='blue',
    linewidth=2
    )
plt.title("MPG change over time")
plt.xlabel("Model Year")
plt.ylabel("MPG")
```

```python
#Create two other visuals or tables related to fuel economy(1).

sns.lineplot(
    data=mpgal, 
  x="cylinders", y="mpg",
    color='red',
    linewidth=2)
plt.title("Cylinders affect on MPG")
plt.xlabel("Cylinders")
plt.ylabel("MPG")

sns.relplot(
    data=mpgal, 
  x="cylinders", y="mpg",
    color='red',
    linewidth=2,
    col="origin",
    kind="line",
)
```

```python
#Create two other visuals or tables related to fuel economy(2).

sns.regplot(
    data=mpgal, 
  x="weight", y="mpg",
)
plt.title("Weight's affect on MPG")
plt.xlabel("Weight")
plt.ylabel("MPG")
```
